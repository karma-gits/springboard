# Capstone Two: Modelling 

### Standarize the datas

```python
from sklearn.preprocessing import StandardScaler
# Selecting relevant features for clustering
X = df[['userId', 'productId','productRating','noOfRatings','price']]

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Applying DBSCAN

```python
from sklearn.cluster import DBSCAN

# Instantiating the DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=5)
# Fitting the model to the scaled data
dbscan.fit(X_scaled)

# Adding a new column 'cluster' to the dataframe with the cluster labels
df['cluster'] = dbscan.labels_

# Function to recommend products for a specific user using KMeans clustering
def recommend_products(user_id):
    user_cluster = df[df['userId'] == user_id]['cluster'].values[0]
    recommended_products = df[df['cluster'] == user_cluster]['productName'].unique()
    return recommended_products

# Example: Recommend products for productId 123 using DBSCAN clustering
user_id = 123
recommended_products = recommend_products(user_id)

print("Recommended products for userId 123 using DBSCAN clustering:")
for product in recommended_products:
    print(product)
```

### Applying KMeans

```python
from sklearn.cluster import KMeans

# Instantiating the KMeans model with the desired number of clusters
kmeans = KMeans(n_clusters=5)

# Fitting the model to the scaled data
kmeans.fit(X_scaled)

# Adding a new column 'cluster' to the dataframe with the cluster labels
df['cluster'] = kmeans.labels_

# Function to recommend products for a specific user using KMeans clustering
def recommend_products(user_id):
    user_cluster = df[df['userId'] == user_id]['cluster'].values[0]
    recommended_products = df[df['cluster'] == user_cluster]['productName'].unique()
    return recommended_products

# Example: Recommend products for productId 123 using KMeans clustering
user_id = 123
recommended_products = recommend_products(user_id)

print("Recommended products for productId 123 using KMeans clustering:")
for product in recommended_products:
    print(product)
```

#### Comparing the metrics

```python
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score

# KMeans 
print('Kmeans ','--'*15)
db_index = davies_bouldin_score(X_scaled, kmeans.labels_)
print("Davies-Bouldin Index:", db_index)
ch_score = calinski_harabasz_score(X_scaled, kmeans.labels_)
print("Calinski-Harabasz Index:", ch_score)

#DBSCAN
print('\nDBSCAN ','--'*15)
db_index = davies_bouldin_score(X_scaled, dbscan.labels_)
print("Davies-Bouldin Index:", db_index)
ch_score = calinski_harabasz_score(X_scaled, dbscan.labels_)
print("Calinski-Harabasz Index:", ch_score)
```

- KMeans has a lower Davies-Bouldin Index, indicating better clustering in terms of cluster separation.
- KMeans has a significantly higher Calinski-Harabasz Index compared to DBSCAN, suggesting that KMeans produces denser and more well-separated clusters.
#### Therefore, KMeans may be considered better for this particular dataset in terms of both cluster separation (Davies-Bouldin Index) and cluster density (Calinski-Harabasz Index).
- ### Since neither model is meeting the desired outcomes, I have decided to proceed with the Collaborative Filtering model.


### Applying the Collaborative filtering process

```python
# Convering data into matrix via pivot
df_matrix= df.pivot_table(index='userId',columns='productName',values='userRating',fill_value=0)
df_matrix.head()
```

### Applying cosine_similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate product similarity matrix
product_similarity_df = pd.DataFrame(cosine_similarity(df_matrix.T), index=df_matrix.columns, columns=df_matrix.columns)

# Function to recommend products to a user
def recommend_products(user_id, n=5):
    user_ratings = df_matrix.loc[user_id] # Filter for user's ratings
    # Calculate the weighted average of similar products >Sort the recommendations in descending order> Exclude products already rated by the user
    recommendations = product_similarity_df.dot(user_ratings).sort_values(ascending=False).drop(df[df['userId'] == user_id]['productName'], errors='ignore')
    return recommendations.head(n) #Get the top n recommendations
```

```python
# 1 Example: Recommend products to user with userId = 123
user_id = 123
recommendations = recommend_products(user_id)
print(recommendations)
```

```python
# 2 Example : Recommend products to user with userId = 200
user_id = 200
recommendations = recommend_products(user_id)
print(recommendations)
```

- ### Given that the quality features consist solely of 'userRating' and 'productId'/'productName', I am of the opinion that collaborative filtering is the most suitable approach in our scenario. I have opted for the cosine similarity method as it is widely employed in collaborative filtering, as it quantifies the resemblance between users or items based on their ratings. Utilizing cosine similarity enables us to pinpoint users who share similar preferences, consequently recommending products they have enjoyed to other users with akin tastes. This strategy aids in delivering personalized recommendations to users predicated on the ratings of like-minded users.
