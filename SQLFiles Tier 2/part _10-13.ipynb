```python
#import libiraries and files
import sqlite3
import pandas as pd

engine = sqlite3.connect('sqlite_db_pythonsqlite.db')
```

""" Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! """

```python
df = pd.read_sql_query(" SELECT f.name, \
                       SUM(CASE WHEN b.memid = 0 THEN f.guestcost * b.slots ELSE f.membercost * b.slots END) AS total_revenue \
                       FROM Facilities AS f JOIN Bookings AS b ON f.facid = b.facid \
                       GROUP BY f.name \
                       HAVING total_revenue < 1000 \
                       ORDER BY total_revenue; ",engine)
df
```

/* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */

```python
df = pd.read_sql_query("SELECT m1.surname AS member_surname, \
       m1.firstname AS member_firstname,\
       m2.surname AS recommender_surname,\
       m2.firstname AS recommender_firstname\
        FROM Members AS m1\
        LEFT JOIN Members AS m2 ON m1.recommendedby = m2.memid\
        ORDER BY m1.surname, m1.firstname; ",engine)
df
```

/* Q12: Find the facilities with their usage by member, but not guests */

```python
df = pd.read_sql_query("SELECT f.name,\
    COUNT(b.memid) AS member_usage \
    FROM Facilities AS f\
    JOIN Bookings AS b ON f.facid = b.facid\
    WHERE b.memid != 0\
    GROUP BY f.name; ",engine)
df
```

/* Q13: Find the facilities usage by month, but not guests */


```python
# >>EXTRACT(MONTH FROM b.starttime<< doesnt work here. so i had to use >> strftime('%m', b.starttime)<<
df = pd.read_sql_query("SELECT f.name,  strftime('%m', b.starttime) AS month, COUNT(b.memid) AS member_usage\
    FROM Facilities AS f\
    JOIN Bookings AS b ON f.facid = b.facid\
    WHERE b.memid != 0\
    GROUP BY f.name,  strftime('%m', b.starttime); ",engine)
df
```
